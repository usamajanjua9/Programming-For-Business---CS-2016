## **Course Description**

This course introduces fundamental programming concepts for **business and FinTech students with no prior programming background**. The course focuses on developing a **problem-solving mindset** and **computational thinking**, enabling students to translate real-world business problems into structured algorithms and programs.

Students will learn core programming constructs using **C++**, including variables, data types, operators, decision-making, loops, functions, arrays, pointers, file handling, and an introductory exposure to **object-oriented programming (OOP)**. Emphasis is placed on **business-oriented applications**, such as financial calculations, reporting, data validation, and automation of repetitive tasks.

The course prepares students to understand how financial and business systems operate internally and provides a strong foundation for advanced courses in **Object-Oriented Programming, Data Analytics, FinTech Systems, and Software-based decision support**.

##  2. Problem-Solving Mindset

###  Programming is NOT Typing Code

Many beginners think:

> Programming means writing code.

 This is **not true**.

 **Programming means solving problems logically**.
Code is just a **tool** to express the solution.

Before writing any code, a good programmer always asks:

* What is the **problem**?
* What is the **goal**?
* What information do I have?
* What result do I want?

---
---

![](https://cdn.dribbble.com/userupload/23005513/file/original-5a9e792c15a007ec1144b5c5befd61e4.gif)

---
###  Understanding the Problem First

Every programming problem has **three core parts**:

####  Input

* Data given to the program
* Example: income, expenses, marks, age

####  Processing (Logic)

* Steps or rules applied to the input
* Example: subtract, compare, calculate percentage

####  Output

* Final result shown to the user
* Example: profit, loss, pass/fail, eligibility

 **No clear input  wrong output**
 **No clear logic  wrong decisions**

---

###  Example: Business Scenario

**Problem:**
A business wants to know whether it is making a **profit or loss**.

Think like a problem solver:

* **Input:**

  * Revenue
  * Cost

* **Logic:**

  * Profit = Revenue  Cost
  * If profit > 0  Profit
  * If profit = 0  Break-even
  * If profit < 0  Loss

* **Output:**

  * Profit value
  * Business status

 Notice:
We solved the problem **without writing any code yet**.

---

###  Breaking a Big Problem into Small Steps

This is called **problem decomposition**.

Instead of thinking:

> I have to write a big program

Think:

1. Take input
2. Do calculation
3. Make a decision
4. Show result

Smaller steps are:

* Easier to understand
* Easier to test
* Easier to fix if something goes wrong

This mindset is **extremely important in FinTech systems**, where mistakes can cost money .

---

###  Common Beginner Mistake

Many beginners:

* Start coding immediately
* Skip thinking
* Get confused when the program doesnt work

 Correct approach:

1. Understand the problem
2. Write steps in plain English
3. Then convert steps into code

>  **Golden Rule:**
> If you cannot explain the solution in words, you cannot code it correctly.

##  3. Computational Thinking

###  How Computers Think

Computers do **exactly what we tell them** - nothing more, nothing less.

They:

* Do **not guess**
* Do **not assume**
* Do **not understand context**

So we must explain everything **clearly and step by step**.

This way of thinking is called **Computational Thinking**.

---

###  What is Computational Thinking?

**Computational Thinking** is a way of solving problems so that:

* A **computer can understand** the solution
* The solution is **clear, logical, and unambiguous**

It helps us convert **real-world problems** into **structured steps**.

---
---

![](https://media0.giphy.com/avatars/RobotHouse/rEsHr1nfaRCz.gif)

---

###  Key Components of Computational Thinking

#### 1 Decomposition (Breaking Down Problems)

Large problems are broken into **smaller parts**.

 Example: Managing expenses

* Take expense input
* Calculate total
* Compare with budget
* Show result

Small problems are easier to solve and test.

---

#### 2 Pattern Recognition

We look for **similarities** in problems.

 Example:

* Profit calculation
* Savings calculation
* Loss calculation

All follow the pattern:

```
result = income  expense
```

Recognizing patterns helps us:

* Reuse logic
* Reduce mistakes
* Write cleaner programs

---

#### 3 Abstraction (Focus on Important Details)

Abstraction means:

* Ignore unnecessary details
* Focus only on what matters

 Example:
When calculating profit, we dont care:

* Who entered the data
* What device was used

We care only about:

* Revenue
* Cost
* Formula

This keeps programs **simple and efficient**.

---

#### 4 Algorithmic Thinking

Algorithmic thinking means:

* Thinking in **ordered steps**
* Each step has a clear purpose
* Steps must be followed in sequence

 Example:

1. Input values
2. Perform calculation
3. Make decision
4. Display output

Computers cannot skip steps  humans can.

---

###  Why Computational Thinking is Important in FinTech

FinTech systems:

* Handle money 
* Process transactions 
* Make decisions automatically 

A small logical mistake can cause:

* Financial loss
* Security issues
* Wrong customer decisions

Computational thinking ensures:

* Correct logic
* Predictable results
* Reliable systems

---

###  Human Thinking vs Computer Thinking

| Human Thinking            | Computer Thinking        |
| ------------------------- | ------------------------ |
| Understands context       | Needs explicit rules     |
| Can guess                 | Cannot guess             |
| Can fix mistakes mentally | Needs exact instructions |
| Flexible                  | Strict                   |

 Thats why **clear thinking** is more important than fast typing.

##  4. Algorithms (Writing Steps in Plain English)

###  What is an Algorithm?

An **algorithm** is a **finite set of clear, ordered steps** used to solve a problem.

In simple words:

> An algorithm is a **recipe** for solving a problem.

Just like a cooking recipe tells:

* what to do first
* what to do next
* when to stop

An algorithm tells the computer **exactly the same way**.

---
---

![](https://cdn-icons-gif.flaticon.com/17122/17122579.gif)

---
###  Characteristics of a Good Algorithm

A good algorithm must be:

*  **Clear**  each step is easy to understand
*  **Ordered**  steps follow a logical sequence
*  **Finite**  it must end after some steps
*  **Correct**  it produces the right output
*  **Language-independent**  can be written without code

At this stage, we **do NOT write code**.
We focus on **thinking and clarity**.

---

###  Writing Algorithms in Plain English

We write algorithms using **simple English sentences**, such as:

* Ask the user to enter
* Calculate
* Check if
* Display the result

This helps beginners:

* Understand the logic
* Avoid syntax confusion
* Translate ideas into code later

---

###  Example 1: Algorithm for Profit or Loss

**Problem:**
Determine whether a business is making profit or loss.

**Algorithm (Plain English):**

1. Start
2. Ask the user to enter **total revenue**
3. Ask the user to enter **total cost**
4. Calculate profit = revenue  cost
5. If profit is greater than 0, display Profit
6. If profit is equal to 0, display Break-even
7. If profit is less than 0, display Loss
8. Display the profit value
9. End

 Notice:

* Steps are simple
* No programming language is used
* Logic is very clear

---

###  Example 2: Algorithm for Student Pass/Fail

**Problem:**
Check whether a student has passed or failed.

**Algorithm (Plain English):**

1. Start
2. Ask the user to enter marks
3. If marks are greater than or equal to 50

   * Display Pass
4. Otherwise

   * Display Fail
5. End

---

###  Common Beginner Mistake

 Writing code directly without an algorithm
 Mixing logic with syntax

 Correct approach:

1. Write algorithm in English
2. Check if it makes sense
3. Then convert it into code

>  If your algorithm is correct, coding becomes easy.

---

###  Why Algorithms Matter in Business & FinTech

Algorithms are used in:

* Loan approval systems
* Fraud detection
* Trading strategies
* Risk analysis
* Credit scoring

Even complex systems are built using **many small algorithms**.

##  5. Flowcharts (Visualizing Algorithms)

###  What is a Flowchart?

A **flowchart** is a **diagrammatic (visual) representation** of an algorithm.

Instead of writing steps in text, we:

* Draw shapes
* Connect them with arrows
* Show how a problem flows from **start to end**

>  Flowcharts help us **see the logic**, not just read it.

They are especially useful for:

* Beginners
* Business discussions
* Explaining logic to non-technical people

---

---
###  Why Use Flowcharts?

Flowcharts help to:

* Understand the **sequence of steps**
* Identify **missing logic**
* Visualize **decision points**
* Reduce logical errors
* Communicate ideas clearly in teams

In business and FinTech, flowcharts are often used during:

* Requirement analysis
* System design meetings
* Process documentation

---

###  Common Flowchart Symbols (Basic Set)

You only need a **few symbols** at this stage.

####  Start / End (Oval)

* Indicates where the process **begins** and **ends**

```
(Start)        (End)
```

---

####  Process (Rectangle)

* Represents a **calculation or action**

```
[ Calculate Profit ]
```

---

####  Decision (Diamond)

* Represents a **yes/no** or **true/false** condition

```
< Profit > 0 >
```

---

####  Flow Lines (Arrows)

* Show the **direction of execution**
* Indicate what happens **next**

---

###  Example: Flowchart for Profit or Loss

Lets visualize the **profit/loss algorithm**.

**Logic Flow:**

Start

Input revenue

Input cost

Calculate profit = revenue  cost

Decision: Is profit > 0?

* Yes  Display Profit
* No  Decision: Is profit = 0?

  * Yes  Display Break-even
  * No  Display Loss
    
    End

 Even without drawing, you can **mentally visualize** the flow.

 Both describe the **same logic**, just in different forms.

##  6. Pseudo-Code (From Diagrams to Structured Steps)

###  What is Pseudo-Code?

**Pseudo-code** is a way of writing program logic using **simple, structured statements** that look like code but are **not tied to any programming language**.

In simple words:

> Pseudo-code is a **bridge** between human thinking and programming code.

It is:

* Easier than real code
* More structured than plain English
* Perfect for beginners

---

###  Why Do We Use Pseudo-Code?

Pseudo-code helps us to:

* Focus on **logic**, not syntax
* Avoid programming errors early
* Plan solutions clearly before coding
* Translate flowcharts into written steps

In business and FinTech projects, pseudo-code is often used during:

* System design
* Logic validation
* Team discussions

---

###  Common Pseudo-Code Keywords

You will often see these simple words:

* `START` / `END`
* `INPUT`
* `OUTPUT`
* `IF`, `ELSE`, `ENDIF`
* `CALCULATE`

These are **not real C++ commands**, just logical placeholders.

---

###  Example 1: Profit or Loss (Pseudo-Code)

```text
START
  INPUT revenue
  INPUT cost
  CALCULATE profit = revenue - cost

  IF profit > 0 THEN
      OUTPUT "Profit"
  ELSE IF profit == 0 THEN
      OUTPUT "Break-even"
  ELSE
      OUTPUT "Loss"
  ENDIF

  OUTPUT profit
END
```

 Notice:

* Logic is clear
* No semicolons
* No data types
* No compiler rules

---

###  Example 2: Pass / Fail Decision

```text
START
  INPUT marks
  IF marks >= 50 THEN
      OUTPUT "Pass"
  ELSE
      OUTPUT "Fail"
  ENDIF
END
```

This pseudo-code can later be **easily converted into C++**.

---

###  Relationship Between Algorithm, Flowchart, and Pseudo-Code

| Representation | Purpose                        |
| -------------- | ------------------------------ |
| Algorithm      | Explain steps in plain English |
| Flowchart      | Visualize the steps            |
| Pseudo-code    | Structure the steps logically  |

 All three describe the **same solution**, just in different formats.

---

###  Common Beginner Mistakes

* Writing full C++ code instead of pseudo-code
* Mixing syntax details (like `;` or `{}`)
* Skipping decision clarity

 Good pseudo-code:

* Is readable
* Is logical
* Is language-independent

##  8. Introduction to C++ Programming

###  What is C++?

**C++** is a **programming language** used to write instructions that a computer can execute.

It is:

* A **compiled language** 
* Used to build **fast and efficient software**
* Common in **finance, trading systems, operating systems**, and performance-critical applications

In this course, C++ is used to:

* Learn **how programming really works**
* Develop **strong logical thinking**
* Build a solid foundation for future courses (especially **OOP next semester**)

---

###  How Does a C++ Program Work?

When you write a C++ program, the computer does **not understand it directly**.

The process is:

1. You write C++ code 
2. A **compiler** converts it into machine code 
3. The machine code is executed by the computer 

This is why C++ is called a **compiled language**.

>  If there is a mistake in the code, the compiler will report an **error**.

---

###  Why Compilation Matters in Business Software

Compiled programs:

* Run **faster**
* Are more **reliable**
* Catch many mistakes **before execution**

In FinTech:

* Speed matters (e.g., trading systems)
* Errors must be caught early
* Reliability is critical when dealing with money 

---

###  First Look at a Simple C++ Program

Lets look at a **very basic C++ program**:

```cpp
#include <iostream>
using namespace std;

int main() {
    cout << "Welcome to Programming for Business!" << endl;
    return 0;
}
```

 Dont worry if you dont understand everything yet.
We will explain each part slowly.

---

###  Key Parts (High-Level View)

* `#include <iostream>`
   Allows input and output operations

* `using namespace std;`
   Makes standard C++ tools easier to use

* `int main()`
   Starting point of every C++ program

* `cout`
   Used to display output

* `return 0;`
   Indicates successful program completion

---

###  Important Rule to Remember

> Every C++ program **must have exactly one `main()` function**.

Execution always starts from `main()`.

---

###  What Happens When You Run the Program?

1. The program starts at `main()`
2. The message is printed on the screen
3. The program ends

Simple, predictable, and controlled  exactly how computers like it.

##  9. Structure of a C++ Program

###  Why Program Structure Matters

A C++ program is not written randomly.
It follows a **specific structure** so that:

* The compiler understands it
* The program runs correctly
* The code is readable and maintainable

Understanding structure early prevents **many beginner errors**.

---

###  Basic Structure of a C++ Program

A simple C++ program has the following main parts:

1. **Header files**
2. **Namespace**
3. **Main function**
4. **Statements inside main**
5. **Return statement**

Lets look at them one by one.

---

###  1. Header Files

```cpp
#include <iostream>
```

* Header files provide **pre-written functionality**
* `<iostream>` allows **input (`cin`) and output (`cout`)**

Think of a header file as:

> A toolbox that gives your program extra tools 

---

###  2. Namespace

```cpp
using namespace std;
```

* `std` stands for **standard**
* It allows us to write `cout` instead of `std::cout`
* Makes code **shorter and cleaner**

 At beginner level, we use it for simplicity.

---

###  3. The `main()` Function

```cpp
int main() {
    // program code
}
```

* `main()` is the **entry point** of the program
* Execution **always starts here**
* Curly braces `{ }` define the program block

>  Without `main()`, the program cannot run.

---

###  4. Statements Inside `main()`

```cpp
cout << "Hello World";
```

* These are **instructions** for the computer
* Executed **line by line**
* Each statement ends with a **semicolon (`;`)**

 Missing semicolons are one of the most common beginner mistakes.

---

###  5. Return Statement

```cpp
return 0;
```

* Tells the system the program ended **successfully**
* `0` means *no error*
* Marks the **end of execution**

---

###  Complete Example with Structure

```cpp
#include <iostream>
using namespace std;

int main() {
    cout << "Welcome to Programming for Business!" << endl;
    return 0;
}
```

---

###  Common Beginner Errors Related to Structure

* Forgetting `#include <iostream>`
* Missing `main()` function
* Missing `{` or `}`
* Forgetting semicolons `;`
* Writing code outside `main()`

Understanding structure helps avoid these errors early.

##  10. Basic Input and Output (`cin` and `cout`)

###  Why Input and Output Are Important

A program is useful only if it can:

* **Take input** from the user 
* **Show output** to the user 

In business and FinTech applications:

* Input = financial data (income, cost, amount, age)
* Output = results (profit, loss, eligibility, balance)

---

###  Output in C++ using `cout`

`cout` is used to **display information on the screen**.

Example:

```cpp
cout << "Hello World";
```

* `cout` sends data to the screen
* `<<` is called the **insertion operator**
* Text must be written inside **double quotes**

---

###  Example: Displaying a Message

```cpp
#include <iostream>
using namespace std;

int main() {
    cout << "Welcome to Programming for Business";
    return 0;
}
```

 When this program runs, the message appears on the screen.

---

###  Input in C++ using `cin`

`cin` is used to **take input from the user**.

Example:

```cpp
cin >> age;
```

* `cin` reads data from the keyboard
* `>>` is called the **extraction operator**
* The input is stored in a **variable**

---

###  Example: Taking Input and Displaying Output

```cpp
#include <iostream>
using namespace std;

int main() {
    int age;

    cout << "Enter your age: ";
    cin >> age;

    cout << "Your age is: " << age;
    return 0;
}
```

 Flow of execution:

1. Program asks for age
2. User enters a number
3. Program prints the entered value

---

###  Business Example: Simple Profit Calculator

```cpp
#include <iostream>
using namespace std;

int main() {
    double revenue, cost, profit;

    cout << "Enter revenue: ";
    cin >> revenue;

    cout << "Enter cost: ";
    cin >> cost;

    profit = revenue - cost;

    cout << "Profit is: " << profit;
    return 0;
}
```

This example shows:

* Input  revenue and cost
* Processing  subtraction
* Output  profit

---

###  Common Beginner Mistakes with `cin` and `cout`

* Forgetting to declare variables before using them
* Using wrong data types (e.g., storing decimals in `int`)
* Forgetting `<<` or `>>`
* Missing semicolons

 These mistakes are normal and part of learning.
