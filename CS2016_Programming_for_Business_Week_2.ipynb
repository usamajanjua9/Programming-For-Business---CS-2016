## ðŸ§  11. Operator Precedence

### ðŸ“Œ What is Operator Precedence?

**Operator precedence** defines the **order in which operations are performed** in an expression.

In simple words:

> Operator precedence tells the computer **which operation to perform first**.

### ðŸ”¢ Order of Execution in Expressions

Just like in mathematics, programming follows an order.

Example:

```cpp
result = 10 + 5 * 2;
```

This is evaluated as:

```
10 + (5 * 2) = 20
```

```
(10 + 5) * 2 = 30
```

ðŸ“Œ Multiplication has **higher precedence** than addition.

---

### ðŸ§® Common Precedence Order (Simplified)

From **highest to lowest priority** (simplified for this course):

1. `()` â€“ Parentheses
2. `*`, `/`, `%` â€“ Multiplication, Division, Modulus
3. `+`, `-` â€“ Addition, Subtraction
4. Relational operators (`>`, `<`, `>=`, `<=`)
5. Equality operators (`==`, `!=`)
6. Logical AND (`&&`)
7. Logical OR (`||`)

---

### ðŸ› ï¸ Using Parentheses to Control Evaluation

Parentheses `()` allow you to **force the order of execution**.

#### ðŸ’¼ Business Example:

```cpp
total = (salary + bonus) * taxRate;
```

Without parentheses:

```cpp
total = salary + bonus * taxRate; // Wrong logic
```

ðŸ“Œ Parentheses make logic:

* Clear
* Safe
* Easy to understand

> ðŸ’¡ Always use parentheses when writing financial expressions.

---

### âš ï¸ Common Logical Errors Due to Incorrect Precedence

âŒ Example:

```cpp
if (income > 50000 && age > 18 || isVIP)
```

This is evaluated as:

```cpp
(income > 50000 && age > 18) || isVIP
```

If the intention was:

```cpp
income > 50000 && (age > 18 || isVIP)
```

Then **parentheses are required**.

---

### ðŸ§  Why Precedence Matters in FinTech

Incorrect precedence can cause:

* Wrong loan approvals
* Incorrect discounts
* Faulty eligibility checks
* Financial losses

In FinTech systems, **logic clarity is critical**.

## ðŸŸ£ Problem 10 â€“ Expression Practice with Precedence

**Task:**
Write a program that:

1. Declares `double salary = 50000;`, `double bonus = 5000;`, `double taxRate = 0.10;`.
2. Calculates **netIncome = (salary + bonus) âˆ’ (salary * taxRate)**.
3. Prints `netIncome`.

Make sure to use **parentheses** correctly.

### âœ… Solution 10

```cpp
#include <iostream>
using namespace std;

int main() {
    double salary = 50000;
    double bonus = 5000;
    double taxRate = 0.10;
    double netIncome;

    netIncome = (salary + bonus) - (salary * taxRate);

    cout << "Net income is: " << netIncome << endl;

    return 0;
}
```
