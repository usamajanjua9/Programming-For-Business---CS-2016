# 1. Introduction to Arrays

---

## 1.1 Meaning of an Array 📦

An **array** is a collection of multiple values of the **same data type** stored under **one name**.

Instead of making many separate variables, we can store many related values inside one array.

### ✅ Simple idea

Suppose you want to store the marks of 5 students.

Without arrays:

```cpp
int mark1 = 78;
int mark2 = 85;
int mark3 = 90;
int mark4 = 67;
int mark5 = 88;
```

With an array:

```cpp
int marks[5] = {78, 85, 90, 67, 88};
```

Now all marks are stored in a single structure called `marks`.

### ✅ Definition

An array is:

* a group of similar values,
* stored in contiguous memory locations,
* accessed using an index number.

### ✅ Real-life analogy

Think of an array like a row of lockers in a school corridor 🏫

* The whole row has one name, for example **LockerRow**
* Each locker has a number: 0, 1, 2, 3, ...
* You can put one item in each locker
* To get an item, you use the locker number

In programming:

* array name = row name
* index = locker number
* value = item inside locker

---

## 1.2 Why Arrays are Useful in Programming 💡

Arrays are very useful because real-world problems usually involve **multiple values**, not just one.

### ✅ Why we need arrays

Imagine a business program where you need to store:

* sales of 30 days,
* prices of 50 products,
* attendance of 40 employees,
* profits of 12 months.

Using separate variables for each value would be difficult and confusing.

### ✅ Arrays help us to:

* store many values easily,
* organize related data,
* use loops for faster processing,
* reduce repetition in code,
* make programs cleaner and shorter,
* perform calculations on a group of values.

### ✅ Example without arrays

```cpp
int sales1 = 1000;
int sales2 = 1200;
int sales3 = 900;
int sales4 = 1500;
int sales5 = 1100;
```

To calculate total sales, you would write:

```cpp
int total = sales1 + sales2 + sales3 + sales4 + sales5;
```

This becomes very long if there are 30 or 100 values.

### ✅ Example with arrays

```cpp
int sales[5] = {1000, 1200, 900, 1500, 1100};
```

Now we can calculate total using a loop:

```cpp
int total = 0;
for (int i = 0; i < 5; i++) {
    total = total + sales[i];
}
```

### ✅ Main benefit

Arrays allow us to work with **many values in a smart and systematic way**.

## 1.4 Difference Between Single Variables and Arrays ⚖️

This is a very important concept.

### ✅ Single Variable

A **single variable** stores only **one value**.

Example:

```cpp
int salary = 50000;
```

This variable holds only one salary.

### ✅ Array

An **array** stores **multiple values** of the same type.

Example:

```cpp
int salaries[5] = {50000, 55000, 60000, 52000, 58000};
```

This array holds 5 salaries.

---

### ✅ Comparison Table

| Feature      | Single Variable                 | Array                      |
| ------------ | ------------------------------- | -------------------------- |
| Stores       | One value                       | Multiple values            |
| Name         | Separate name for each variable | One common name            |
| Access       | Directly by variable name       | By index number            |
| Suitable for | One item only                   | Many related items         |
| Example      | One employee salary             | Salaries of many employees |

---

### ✅ Example using single variables

```cpp
int mark1 = 78;
int mark2 = 85;
int mark3 = 90;
```

### ✅ Example using array

```cpp
int marks[3] = {78, 85, 90};
```

### ✅ Which is better?

* If you only need one value, a simple variable is enough.
* If you need many related values, an array is much better.

### ✅ Key point

Use:

* **single variable** for one item,
* **array** for many similar items.

# 2. One-Dimensional Arrays

---

## 2.1 Concept of One-Dimensional Arrays 📏

A **one-dimensional array** is the simplest type of array.
It stores values in a **single line** or **single sequence**.

That means:

* values are arranged one after another,
* each value has one position,
* each position is identified by one index.

### ✅ Example

```cpp
int scores[5] = {10, 20, 30, 40, 50};
```

This is a one-dimensional array because it is a simple list of values.

### ✅ Visual idea

| Index | 0  | 1  | 2  | 3  | 4  |
| ----- | -- | -- | -- | -- | -- |
| Value | 10 | 20 | 30 | 40 | 50 |

### ✅ Why called one-dimensional?

Because only **one index** is needed to access each value.

For example:

* `scores[0]` → 10
* `scores[1]` → 20
* `scores[2]` → 30

### ✅ Business example

```cpp
int monthlyExpenses[6] = {2000, 2500, 1800, 3000, 2700, 2200};
```

This stores expenses for 6 months in one sequence.

## 2.2 Declaring an Array ✍️

Before using an array, we must **declare** it.

### ✅ Syntax

```cpp
dataType arrayName[size];
```

### ✅ Parts of declaration

* `dataType` → type of values (`int`, `float`, `char`, etc.)
* `arrayName` → name of the array
* `size` → number of elements the array can store

### ✅ Examples

```cpp
int marks[5];
float prices[10];
char grades[4];
```

### ✅ Meaning of these declarations

* `marks[5]` → can store 5 integer values
* `prices[10]` → can store 10 decimal values
* `grades[4]` → can store 4 characters

### ✅ Business examples

```cpp
int sales[7];
float productPrices[20];
int employeeAges[15];
```

These arrays can be used to store:

* 7 days of sales,
* 20 product prices,
* 15 employee ages.

### ✅ Important note

The size should be decided when declaring the array in basic C++ programs.

## 2.3 Initializing an Array 🚀

**Initialization** means giving values to the array at the time of declaration.

### ✅ Syntax

```cpp
dataType arrayName[size] = {value1, value2, value3, ...};
```

### ✅ Example

```cpp
int marks[5] = {78, 85, 90, 67, 88};
```

This creates an array and stores 5 values in it.

### ✅ Another example

```cpp
float prices[4] = {99.5, 120.0, 150.75, 80.25};
```

### ✅ Partial initialization

```cpp
int numbers[5] = {10, 20};
```

Here:

* first value = 10
* second value = 20
* remaining values become 0

### ✅ Output idea

| Index | 0  | 1  | 2 | 3 | 4 |
| ----- | -- | -- | - | - | - |
| Value | 10 | 20 | 0 | 0 | 0 |

### ✅ Business example

```cpp
int weeklySales[7] = {1200, 1400, 1600, 1800, 1500, 1700, 2000};
```

This stores the sales of one week.

### ✅ Why initialization is helpful

It saves time and makes the program easier to read when values are already known.

## 2.4 Accessing Array Elements 🔍

To use a value from an array, we access it through its **index**.

### ✅ Syntax

```cpp
arrayName[index]
```

### ✅ Example

```cpp
int marks[5] = {78, 85, 90, 67, 88};
cout << marks[0];
cout << marks[2];
```

### ✅ Output

* `marks[0]` → 78
* `marks[2]` → 90

### ✅ Business example

```cpp
float prices[4] = {99.5, 120.0, 150.75, 80.25};
cout << prices[1];
```

Output:

```cpp
120.0
```

### ✅ Important rule

Array indexing starts from **0**, not from 1.

So in an array of size 5:

* first element is at index `0`
* last element is at index `4`

### ✅ Example with loop

```cpp
int sales[5] = {1000, 1200, 1400, 1600, 1800};

for (int i = 0; i < 5; i++) {
    cout << sales[i] << endl;
}
```

This prints all elements one by one.

## 2.6 Array Index and Position 🔢

This topic is very important because many students get confused between **index** and **position**.

### ✅ Position

Position means the natural counting order:

* 1st item
* 2nd item
* 3rd item

### ✅ Index

Index is the array number used in programming:

* first item → index 0
* second item → index 1
* third item → index 2

### ✅ Example

```cpp
int data[4] = {10, 20, 30, 40};
```

| Position | 1st | 2nd | 3rd | 4th |
| -------- | --- | --- | --- | --- |
| Index    | 0   | 1   | 2   | 3   |
| Value    | 10  | 20  | 30  | 40  |

### ✅ Important rule

For an array of size `n`:

* first index = `0`
* last index = `n - 1`

### ✅ Example

If array size is 6:

* first index = 0
* last index = 5

### ✅ Common mistake

Students often try:

```cpp
int marks[5] = {10, 20, 30, 40, 50};
cout << marks[5];
```

This is wrong ❌ because valid indexes are:
`0, 1, 2, 3, 4`

There is no element at index `5`.

## 2.7 Array Size and Memory Idea 💾

An array has a **fixed size** in basic C++.

That means when you declare it, you decide how many elements it can store.

### ✅ Example

```cpp
int marks[5];
```

This array can store exactly **5 integers**.

Not:

* 4 only,
* not 6,
* exactly 5 spaces are reserved.

### ✅ Memory idea

When an array is created, memory is allocated for all elements together in a continuous block.

That is why arrays are efficient and easy to process with loops.

### ✅ Simple concept

If one integer takes 4 bytes, then:

```cpp
int marks[5];
```

may use:

```cpp
5 × 4 = 20 bytes
```

### ✅ Why this matters

If you declare a very large array, it uses more memory.

### ✅ Business example

```cpp
float salaries[100];
```

This reserves space for 100 salary values.

### ✅ Fixed size limitation

If you create:

```cpp
int sales[5];
```

you cannot store a 6th value in it.

So it is important to choose array size carefully.

### ✅ Key learning

Array size:

* tells how many values can be stored,
* is fixed in simple programs,
* affects memory usage,
* helps control the structure of business records.

# 3. Taking Input and Displaying Output with Arrays

When we use arrays in real programs, we usually need to:

* **take multiple values as input** from the user,
* **store them in an array**,
* and then **display them properly**.

This is one of the biggest advantages of arrays:
instead of writing many separate input and output statements, we can use **loops** 🔁

---

## 3.1 Entering Data into an Array ⌨️

We can enter values into an array one by one using a loop.

### ✅ General idea

If an array has 5 elements, we can ask the user to enter 5 values and store them at indexes:

* `0`
* `1`
* `2`
* `3`
* `4`

### ✅ Example: Input marks of 5 students

```cpp
#include <iostream>
using namespace std;

int main() {
    int marks[5];

    for (int i = 0; i < 5; i++) {
        cout << "Enter marks for student " << i + 1 << ": ";
        cin >> marks[i];
    }

    return 0;
}
```

### ✅ What happens here?

* `marks[5]` creates an array of 5 integers
* loop runs 5 times
* each time, one value is entered
* value is stored in `marks[i]`

### ✅ Example run

```cpp
Enter marks for student 1: 78
Enter marks for student 2: 85
Enter marks for student 3: 90
Enter marks for student 4: 67
Enter marks for student 5: 88
```

### ✅ Business example: Input daily sales

```cpp
#include <iostream>
using namespace std;

int main() {
    int sales[7];

    for (int i = 0; i < 7; i++) {
        cout << "Enter sales for day " << i + 1 << ": ";
        cin >> sales[i];
    }

    return 0;
}
```

This program stores sales of 7 days in an array.

---

## 3.2 Using Loops with Arrays 🔁

Arrays and loops are best friends in programming 😄

A loop helps us:

* enter values,
* display values,
* calculate totals,
* search data,
* update data.

### ✅ Why loops are important

Imagine storing 100 product prices.

Without a loop, you would need 100 input statements:

```cpp
cin >> prices[0];
cin >> prices[1];
cin >> prices[2];
...
```

That would be too long and inefficient ❌

With a loop:

```cpp
for (int i = 0; i < 100; i++) {
    cin >> prices[i];
}
```

Much easier ✅

### ✅ Example: Input and display values

```cpp
#include <iostream>
using namespace std;

int main() {
    int numbers[5];

    for (int i = 0; i < 5; i++) {
        cout << "Enter value " << i + 1 << ": ";
        cin >> numbers[i];
    }

    cout << "\nYou entered:\n";
    for (int i = 0; i < 5; i++) {
        cout << numbers[i] << endl;
    }

    return 0;
}
```

### ✅ Key point

Whenever we work with arrays, we usually use:

* **`for` loop** for fixed number of elements,
* sometimes **`while` loop** if needed.

But for arrays, `for` loop is most common.

---

## 3.3 Printing Array Contents 🖥️

After storing data in an array, we often want to display it.

### ✅ Syntax idea

We print each element one by one using a loop.

### ✅ Example

```cpp
#include <iostream>
using namespace std;

int main() {
    int marks[5] = {78, 85, 90, 67, 88};

    cout << "Student Marks:\n";
    for (int i = 0; i < 5; i++) {
        cout << "Student " << i + 1 << ": " << marks[i] << endl;
    }

    return 0;
}
```

### ✅ Output

```cpp
Student Marks:
Student 1: 78
Student 2: 85
Student 3: 90
Student 4: 67
Student 5: 88
```

### ✅ Business example: Display product prices

```cpp
#include <iostream>
using namespace std;

int main() {
    float prices[4] = {99.5, 120.0, 150.75, 80.25};

    cout << "Product Prices:\n";
    for (int i = 0; i < 4; i++) {
        cout << "Product " << i + 1 << ": " << prices[i] << endl;
    }

    return 0;
}
```

### ✅ Why output formatting matters

Good display makes the program:

* easier to understand,
* more professional,
* more useful for reports.

## 3.4 Common Input and Output Mistakes ⚠️

Students often make small mistakes while using arrays. Let’s look at the common ones.

### ❌ Mistake 1: Wrong loop limit

```cpp
for (int i = 0; i <= 5; i++)
```

If array size is 5, valid indexes are:
`0, 1, 2, 3, 4`

So `i <= 5` is wrong because it also tries index `5`.

✅ Correct:

```cpp
for (int i = 0; i < 5; i++)
```

---

### ❌ Mistake 2: Forgetting to use index

```cpp
cin >> marks;
```

This is wrong because `marks` is the whole array, not a single element.

✅ Correct:

```cpp
cin >> marks[i];
```

---

### ❌ Mistake 3: Printing only one element by mistake

```cpp
cout << marks[0];
```

This prints only the first value, not the whole array.

✅ Correct:
Use a loop to print all values.

---

### ❌ Mistake 4: Using invalid index

```cpp
cout << marks[10];
```

If array size is 5, index 10 is invalid.

---

### ❌ Mistake 5: Inputting fewer or extra values

If loop runs wrong number of times, data will be incomplete or invalid.

### ✅ Best practice

Always check:

* array size,
* loop condition,
* input statement,
* output statement.

# 5. Analyzing Data Stored in Arrays

Storing data is only the first step.
The real power of programming comes when we **analyze the data** 🔍

Using arrays, we can:

* find total,
* calculate average,
* find highest value,
* find lowest value,
* count certain records.

This is very useful in business and finance.

---

## 5.1 Finding Total ➕

We often need the total of all values in an array.

### ✅ Example: Total weekly sales

```cpp
#include <iostream>
using namespace std;

int main() {
    int sales[5] = {1000, 1200, 900, 1500, 1100};
    int total = 0;

    for (int i = 0; i < 5; i++) {
        total = total + sales[i];
    }

    cout << "Total Sales = " << total << endl;

    return 0;
}
```

### ✅ Output

```cpp
Total Sales = 5700
```

### ✅ Business importance

Total helps in:

* calculating total revenue,
* total expenses,
* total attendance,
* total stock sold.

---

## 5.2 Finding Average 📈

Average gives a general idea of performance.

### ✅ Formula

$
\text{Average} = \frac{\text{Total}}{\text{Number of values}}
$

### ✅ Example

```cpp
#include <iostream>
using namespace std;

int main() {
    int marks[5] = {78, 85, 90, 67, 88};
    int total = 0;
    float average;

    for (int i = 0; i < 5; i++) {
        total = total + marks[i];
    }

    average = total / 5.0;

    cout << "Average Marks = " << average << endl;

    return 0;
}
```

### ✅ Business example

Average can be used for:

* average daily sales,
* average employee attendance,
* average product rating,
* average monthly profit.

---

## 5.3 Finding Maximum Value 🔺

Maximum means the largest value in the array.

### ✅ Example

```cpp
#include <iostream>
using namespace std;

int main() {
    int sales[5] = {1000, 1200, 900, 1500, 1100};
    int max = sales[0];

    for (int i = 1; i < 5; i++) {
        if (sales[i] > max) {
            max = sales[i];
        }
    }

    cout << "Highest Sales = " << max << endl;

    return 0;
}
```

### ✅ Output

```cpp
Highest Sales = 1500
```

### ✅ Business importance

Maximum helps identify:

* highest sales day,
* highest salary,
* highest expense,
* most expensive product.

---

## 5.4 Finding Minimum Value 🔻

# 6. Searching in Arrays

In many programs, storing data is not enough.
We also want to **find a specific value** inside the array.

This process is called **searching** 🔍

For example, in a business program, we may want to:

* find whether a product price exists,
* find a specific sales value,
* check whether an employee ID is present,
* search for a student's marks,
* locate a certain stock quantity.

---

## 6.1 Meaning of Searching 🔎

**Searching** means checking whether a required value exists in an array or not.

### ✅ Simple idea

Suppose we have this array:

```cpp id="p2s1a1"
int sales[5] = {1000, 1200, 900, 1500, 1100};
```

Now if we want to know whether `900` is present, we must compare `900` with each element of the array.

If found, we say:

* the value exists ✅

If not found, we say:

* the value does not exist ❌

### ✅ Why searching is important

Searching helps in:

* checking records,
* verifying entries,
* locating data for editing,
* finding whether an item exists before processing it.

### ✅ Business examples

Searching can be used to:

* find whether a product with a certain price exists,
* search for a customer's bill amount,
* locate a salary value,
* check whether sales target was achieved on a day.

---

## 6.2 Linear Search Concept 📍

The simplest searching method for arrays is called **Linear Search**.

### ✅ What is linear search?

In linear search, we start from the first element and check one by one until:

* the value is found, or
* the array ends.

### ✅ How it works

For an array:

```cpp id="f2y9b2"
int numbers[5] = {10, 20, 30, 40, 50};
```

If we want to search for `30`, we check:

* Is `numbers[0] == 30`? → No
* Is `numbers[1] == 30`? → No
* Is `numbers[2] == 30`? → Yes ✅

So value is found at index `2`.

### ✅ Why called “linear”?

Because we move in a straight line through the array, from start to end.

### ✅ When is linear search useful?

Linear search is useful when:

* array is small,
* data is unsorted,
* we want a simple beginner-friendly solution.

# 7. Common Errors in Arrays

Arrays are very useful, but beginners often make mistakes while using them.
Understanding these errors helps students write better programs and debug easily 🛠️

---

## 7.1 Index Out of Range Problems 🚫

This is one of the most common array mistakes.

### ✅ What does it mean?

It means trying to access an index that does not exist.

### ✅ Example

```cpp id="g2u4k9"
int marks[5] = {10, 20, 30, 40, 50};
cout << marks[5];
```

This is wrong ❌ because valid indexes are:

* 0
* 1
* 2
* 3
* 4

There is no index `5`.

### ✅ Correct understanding

For array of size `n`, valid indexes are:
[
0 \text{ to } n-1
]

### ✅ Why dangerous?

It may:

* show garbage value,
* cause unexpected behavior,
* crash the program in some cases.

---

## 7.2 Using Wrong Loop Limits 🔁

Another common problem is using the wrong loop condition.

### ❌ Wrong example

```cpp id="n4v7l1"
for (int i = 0; i <= 5; i++) {
    cout << marks[i] << endl;
}
```

If array size is 5, then `i <= 5` is wrong because it includes index `5`.

### ✅ Correct example

```cpp id="q6w8z2"
for (int i = 0; i < 5; i++) {
    cout << marks[i] << endl;
}
```

### ✅ Rule

Use:

```cpp id="r8x9y3"
i < size
```

not

```cpp id="s1c2v4"
i <= size
```

### ✅ Why this matters

Wrong loop limits can cause:

* extra input,
* invalid output,
* access to out-of-range elements.

---

## 7.3 Uninitialized Array Values ⚠️

If an array is declared but values are not assigned, then its elements may contain garbage values.

### ✅ Example

```cpp id="d3f5g5"
int data[5];
cout << data[0];
```

This may print an undefined or garbage value.

### ✅ Why?

Because memory was reserved, but no proper value was stored yet.

### ✅ Better approach

Always initialize or take input before using array values.

### ✅ Correct example

```cpp id="h6j7k6"
int data[5] = {0, 0, 0, 0, 0};
```

or

```cpp id="l8m9n7"
for (int i = 0; i < 5; i++) {
    cin >> data[i];
}
```

### ✅ Lesson

Never use array elements before assigning values.

---

## 7.4 Input Handling Mistakes ⌨️

Students also make mistakes while taking input.

### ❌ Mistake 1: Forgetting index

```cpp id="p1q2r8"
cin >> marks;
```

Wrong ❌ because `marks` is the whole array.

✅ Correct:

```cpp id="s3t4u9"
cin >> marks[i];
```

---

### ❌ Mistake 2: Taking fewer inputs than needed

If array size is 5 but loop runs only 3 times, then remaining values are not properly assigned.

### ❌ Mistake 3: Taking too many inputs

If array size is 5 but loop runs 6 times, program tries to store an extra value outside the array.

### ✅ Correct input style

```cpp id="v5w6x0"
for (int i = 0; i < 5; i++) {
    cin >> marks[i];
}
```

### ✅ Best practice

Always match:

* array size
* loop limit
* number of inputs

---

# 8. Practical Class Activities

Now let’s convert the concepts into classroom tasks 🎯
These activities help students practice arrays in real and meaningful ways.

---

## 8.1 Store and Display Student Marks 🎓

### ✅ Task

Ask students to:

* create an array of 5 marks,
* take input from the user,
* display all marks.

### ✅ Sample code

```cpp id="a1b2c1"
#include <iostream>
using namespace std;

int main() {
    int marks[5];

    for (int i = 0; i < 5; i++) {
        cout << "Enter marks for student " << i + 1 << ": ";
        cin >> marks[i];
    }

    cout << "\nStudent Marks:\n";
    for (int i = 0; i < 5; i++) {
        cout << "Student " << i + 1 << ": " << marks[i] << endl;
    }

    return 0;
}
```

## 2.2 Passing Arrays to Functions 🚚

When we want a function to work on an array, we pass the array as an argument.

### ✅ Basic syntax

```cpp
returnType functionName(dataType arrayName[], int size)
```

### ✅ Example

```cpp
void display(int arr[], int size)
```

This means:

* function name = `display`
* it receives an integer array
* it also receives the size of the array

### ✅ Why pass size too?

Because the function needs to know how many elements are in the array.

### ✅ Example program

```cpp
#include <iostream>
using namespace std;

void display(int arr[], int size) {
    for (int i = 0; i < size; i++) {
        cout << arr[i] << " ";
    }
    cout << endl;
}

int main() {
    int marks[5] = {78, 85, 90, 67, 88};
    display(marks, 5);

    return 0;
}
```

### ✅ Output

```cpp
78 85 90 67 88
```

## 2.3 Functions for Input ⌨️

* receives the array,
* receives its size,

### ✅ Business example

### ✅ Why this is useful

Whenever we need to display the same kind of data again, we can simply call the function.

## 2.5 Functions for Calculations 🧮

Functions can also perform calculations on arrays, such as finding total or average.

### ✅ Example: Function to calculate total

```cpp
#include <iostream>
using namespace std;

int calculateTotal(int arr[], int size) {
    int total = 0;

    for (int i = 0; i < size; i++) {
        total = total + arr[i];
    }

    return total;
}

int main() {
    int sales[5] = {1000, 1200, 900, 1500, 1100};
    int totalSales = calculateTotal(sales, 5);

    cout << "Total Sales = " << totalSales << endl;

    return 0;
}
```

### ✅ Example: Function to calculate average

```cpp
#include <iostream>
using namespace std;

float calculateAverage(int arr[], int size) {
    int total = 0;

    for (int i = 0; i < size; i++) {
        total = total + arr[i];
    }

    return total / (float)size;
}

int main() {
    int marks[5] = {78, 85, 90, 67, 88};

    cout << "Average Marks = " << calculateAverage(marks, 5) << endl;

    return 0;
}
```

### ✅ Business use

Calculation functions are useful for:

* total sales
* average salary
* total expenses
* average rating
* monthly profits

## 2.6 Functions for Searching and Analysis 🔍

We can also create functions to search arrays or find maximum/minimum values.

### ✅ Example: Search function

```cpp
#include <iostream>
using namespace std;

bool searchValue(int arr[], int size, int target) {
    for (int i = 0; i < size; i++) {
        if (arr[i] == target) {
            return true;
        }
    }
    return false;
}

int main() {
    int sales[5] = {1000, 1200, 900, 1500, 1100};
    int target;

    cout << "Enter value to search: ";
    cin >> target;

    if (searchValue(sales, 5, target)) {
        cout << "Value found.\n";
    } else {
        cout << "Value not found.\n";
    }

    return 0;
}
```

### ✅ Example: Maximum function

```cpp
#include <iostream>
using namespace std;

int findMax(int arr[], int size) {
    int max = arr[0];

    for (int i = 1; i < size; i++) {
        if (arr[i] > max) {
            max = arr[i];
        }
    }

    return max;
}

int main() {
    int profits[5] = {5000, 7000, 6500, 8000, 7200};

    cout << "Highest Profit = " << findMax(profits, 5) << endl;

    return 0;
}
```

### ✅ Why analysis functions matter

These functions allow us to separate logic clearly:

* input function → stores data
* display function → prints data
* calculation function → computes results
* search function → finds data

This is the beginning of **modular programming**.

### ✅ Why modular programs are better

* easier to expand later.

### ✅ Example structure of a modular program

A business record program may have:

* `findMax()` → find highest sales

### ✅ Comparison

| Non-Modular Program      | Modular Program             |
| ------------------------ | --------------------------- |

| Harder to read           | Easier to understand        |
| Difficult to debug       | Easier to fix               |

### ✅ Key lesson

## 7.1 Incorrect Function Use with Arrays 🛠️

Sometimes students create functions for arrays but pass the wrong parameters.

### ❌ Example

```cpp id="w11e71"
void display(int arr) {
    cout << arr;
}
```

This is wrong because `arr` should be an array, not a single integer.

### ✅ Correct version

```cpp id="w11e72"
void display(int arr[], int size) {
    for (int i = 0; i < size; i++) {
        cout << arr[i] << endl;
    }
}
```

### ✅ Lesson

When passing arrays to functions:

* pass the array,
* also pass its size.

## 7.2 Input Problems in Strings ⌨️

A common beginner issue is using `cin` when the input contains spaces.

### ❌ Example

```cpp id="w11e73"
string fullName;
cin >> fullName;
```

If user enters:

```cpp id="w11e74"
Ali Khan
```

only `Ali` is stored.

### ✅ Better solution

```cpp id="w11e75"
getline(cin, fullName);
```

### ✅ Lesson

Use:

* `cin` for one-word input
* `getline()` for full-line input

---

## 7.3 Mixing `cin` and `getline()` Issues 🔄

This is one of the most common string input problems.

### ✅ What happens?

If `cin` is used before `getline()`, the leftover newline may cause `getline()` to read an empty line.

### ❌ Example

```cpp id="w11e76"
int age;
string name;

cin >> age;
getline(cin, name);
```

Here `getline()` may not work properly because it reads the leftover Enter key.

### ✅ Correct approach

Use `cin.ignore()` before `getline()`.

```cpp id="w11e77"
int age;
string name;

cin >> age;
cin.ignore();
getline(cin, name);
```

### ✅ Why this matters

Without this fix, students may get:

* empty string input,
* skipped input,
* confusing behavior.

---

## 7.4 Poorly Formatted Console Output 📄

Sometimes the logic is correct, but the output looks messy.

### ❌ Example

```cpp id="w11e78"
cout << name << salary << department;
```

This may produce:

```cpp id="w11e79"

```

Very difficult to read ❌

### ✅ Better version

```cpp id="w11e80"
cout << "Name       : " << name << endl;
cout << "Salary     : " << salary << endl;
cout << "Department : " << department << endl;
```

### ✅ Lesson

Always use:

* line breaks,
* headings.

# Multidimensional Arrays

A **multidimensional array** is an array that has **more than one dimension**.
The most common type is the **two-dimensional array**, which works like a **table** made of rows and columns.

It is useful when data is naturally arranged in a grid, such as:

* student marks in different subjects
* monthly sales for multiple branches
* seating arrangement in a classroom
* matrix operations in mathematics

---

# 1. What is a Two-Dimensional Array?

A normal array stores data in a single line:

```cpp
int marks[5];
```

A two-dimensional array stores data in the form of **rows and columns**:

```cpp
int marks[3][4];
```

This means:

* `3` rows
* `4` columns

So this array can store:

**3 × 4 = 12 values**

You can imagine it like this:

| Row \ Column |  0 |  1 |  2 |  3 |
| ------------ | -: | -: | -: | -: |
| 0            |    |    |    |    |
| 1            |    |    |    |    |
| 2            |    |    |    |    |

---

# 2. Syntax of a Two-Dimensional Array

```cpp
dataType arrayName[rows][columns];
```

Example:

```cpp
int matrix[2][3];
```

This creates a 2D array with:

* 2 rows
* 3 columns

---

# 3. Accessing Elements in a 2D Array

To access an element, we use:

```cpp
arrayName[row][column]
```

Example:

```cpp
matrix[0][0] = 10;
matrix[0][1] = 20;
matrix[1][2] = 30;
```

Explanation:

* `matrix[0][0]` → first row, first column
* `matrix[0][1]` → first row, second column
* `matrix[1][2]` → second row, third column

---

# 4. Example of Declaring and Assigning Values

```cpp
#include <iostream>
using namespace std;

int main() {
    int marks[2][3] = {
        {85, 90, 78},
        {88, 76, 92}
    };

    cout << "First student's first subject marks: " << marks[0][0] << endl;
    cout << "Second student's third subject marks: " << marks[1][2] << endl;

    return 0;
}
```

### Output:

```cpp
First student's first subject marks: 85
Second student's third subject marks: 92
```

---

# 5. Input in a Two-Dimensional Array

Usually, **nested loops** are used to take input.

```cpp
#include <iostream>
using namespace std;

int main() {
    int arr[2][3];

    for (int i = 0; i < 2; i++) {
        for (int j = 0; j < 3; j++) {
            cout << "Enter value for arr[" << i << "][" << j << "]: ";
            cin >> arr[i][j];
        }
    }

    return 0;
}
```

### Explanation:

* Outer loop controls the **rows**
* Inner loop controls the **columns**

---

# 6. Displaying a Two-Dimensional Array

```cpp
#include <iostream>
using namespace std;

int main() {
    int arr[2][3] = {
        {1, 2, 3},
        {4, 5, 6}
    };

    cout << "Array elements are:\n";

    for (int i = 0; i < 2; i++) {
        for (int j = 0; j < 3; j++) {
            cout << arr[i][j] << " ";
        }
        cout << endl;
    }

    return 0;
}
```

### Output:

```cpp
Array elements are:
1 2 3
4 5 6
```

```cpp
int arr[2][3][4];
```

This means:

* 2 blocks
* 3 rows
* 4 columns

These are less common for beginners, but they are useful for things like:

* storing data for multiple classes, rows, and columns

* image or scientific data

Example declaration:

```cpp
int data[2][2][2] = {
    {
        {1, 2},
        {3, 4}
    },
    {
        {5, 6},
        {7, 8}
    }
};
```

# 10. Important Points

* Indexing starts from **0**
* A 2D array is accessed using **two indexes**
* Nested loops are normally used with multidimensional arrays
* Rows and columns must be defined properly
* Multidimensional arrays are best for **table-like data**

# 11. Advantages of Multidimensional Arrays

* organize data in table form
* easy to represent rows and columns
* useful in matrices, reports, and business records
* allows structured storage of related data

# 12. Limitations

* fixed size
* can become harder to manage if dimensions increase
* wastes memory if many positions remain unused

Tic-Tac-Toe (Tix Cross)** in **C++**

```cpp
#include <iostream>                           // For input and output
using namespace std;                          // To use cout and cin easily

int main()                                    // Main function starts
{
    char box1 = '1', box2 = '2', box3 = '3';  // First row boxes
    char box4 = '4', box5 = '5', box6 = '6';  // Second row boxes
    char box7 = '7', box8 = '8', box9 = '9';  // Third row boxes

    int choice;                               // To store user input
    char player = 'X';                        // First player is X
    int moves = 0;                            // Count total moves
    bool win = false;                         // To check if someone wins

    cout << "===== TIC TAC TOE GAME =====" << endl;     // Game title
    cout << "Player 1 = X" << endl;                     // Show player 1 symbol
    cout << "Player 2 = O" << endl;                     // Show player 2 symbol
    cout << "Choose box number from 1 to 9" << endl;   // Instructions

    while (win == false && moves < 9)         // Run until win or all 9 moves used
    {
        cout << endl;
        cout << " " << box1 << " | " << box2 << " | " << box3 << endl;
        cout << "---|---|---" << endl;
        cout << " " << box4 << " | " << box5 << " | " << box6 << endl;
        cout << "---|---|---" << endl;
        cout << " " << box7 << " | " << box8 << " | " << box9 << endl;
        cout << endl;

        cout << "Player " << player << ", enter box number: ";   // Ask current player
        cin >> choice;                                            // Take input

        if (choice == 1 && box1 != 'X' && box1 != 'O') box1 = player;
        else if (choice == 2 && box2 != 'X' && box2 != 'O') box2 = player;
        else if (choice == 3 && box3 != 'X' && box3 != 'O') box3 = player;
        else if (choice == 4 && box4 != 'X' && box4 != 'O') box4 = player;
        else if (choice == 5 && box5 != 'X' && box5 != 'O') box5 = player;
        else if (choice == 6 && box6 != 'X' && box6 != 'O') box6 = player;
        else if (choice == 7 && box7 != 'X' && box7 != 'O') box7 = player;
        else if (choice == 8 && box8 != 'X' && box8 != 'O') box8 = player;
        else if (choice == 9 && box9 != 'X' && box9 != 'O') box9 = player;
        else
        {
            cout << "Invalid move! Box already filled or wrong number." << endl;
            continue;                                             // Ask again
        }

        moves++;                                                  // Increase move count

        if ((box1 == box2 && box2 == box3) ||                    // Row 1
            (box4 == box5 && box5 == box6) ||                    // Row 2
            (box7 == box8 && box8 == box9) ||                    // Row 3
            (box1 == box4 && box4 == box7) ||                    // Column 1
            (box2 == box5 && box5 == box8) ||                    // Column 2
            (box3 == box6 && box6 == box9) ||                    // Column 3
            (box1 == box5 && box5 == box9) ||                    // Diagonal 1
            (box3 == box5 && box5 == box7))                      // Diagonal 2
        {
            win = true;                                          // Someone won
        }
        else
        {
            if (player == 'X')                                   // Change player
                player = 'O';
            else
                player = 'X';
        }
    }

    cout << endl;
    cout << "Final Board:" << endl;
    cout << " " << box1 << " | " << box2 << " | " << box3 << endl;
    cout << "---|---|---" << endl;
    cout << " " << box4 << " | " << box5 << " | " << box6 << endl;
    cout << "---|---|---" << endl;
    cout << " " << box7 << " | " << box8 << " | " << box9 << endl;
    cout << endl;

    if (win == true)
        cout << "Player " << player << " wins the game!" << endl;
    else
        cout << "Game is a draw!" << endl;

    return 0;                                                    // End program
}
```

### Why this version is easier

* no functions
* all logic in `main()`
* easy box-by-box system
* simple for beginners
* directly playable

### Box positions

```cpp
 1 | 2 | 3
---|---|---
 4 | 5 | 6
---|---|---
 7 | 8 | 9
```

### Example play

* Player X chooses `5`
* Player O chooses `1`
* Player X chooses `9`

That fills those box numbers with `X` and `O`.
